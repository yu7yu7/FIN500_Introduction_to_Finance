Question 1: A stock's rate of return standard deviation is estimated to be $\sigma$ equals 0.35, and the continuously-compounded riskfree interest rate is $r$ = 3.00%. This information is being used to calibrate the parameters u, d, and Rf for the binomial model with a period of $\Delta t$ equals $1/12$ (one month).  What is the binomial model's risk-neutral probability that the stock returns u over a period, $\hat{\pi}$?

```python
import math as m
σ = 0.35
Δt = 1/12
r = 0.03

u = m.exp(σ* m.sqrt(Δt))
d = 1/u
Rf = m.exp(r * Δt)

π̂ = (Rf - d)/(u - d)
π̂
```

Question 2: An American put option has an exercise price of $\$ 50$ and is written on a stock that is currently worth $S_t$ = $\$ 45$. Based on the binomial model with a period of $\Delta t$ =1/52 (one week), the value of the put next period is $P_{t+\Delta t,u} = 3$ if the stock returns $u$ and  $P_{t+\Delta t,d} = 6$ if the stock returns $d$. Furthermore, the risk-neutral probability that the stock returns $u$ equals $\hat{\pi}$ equals $0.5$, and the continuously-compounded riskfree rate equals $r = 4.00 \%$. What is the put option's current value, $P_t$ ?

```python
import jax
import jax.numpy as jnp
import math

def Pt(X, St, r, Δt, π̂, Pu, Pd):
    return max(X - St, math.exp(-1*r*Δt)* (π̂ * Pu + (1-π̂)*Pd))
print(Pt(50, 45, 0.04,1/52, 0.5, 3, 6))
```
