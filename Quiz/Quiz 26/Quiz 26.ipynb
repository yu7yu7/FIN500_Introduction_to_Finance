# Quiz 1

A firm has risky assets worth $A0 = \$10$ million that have a rate of return volatility of $σ = 0.40$. The firm pays out dividends at the rate of $δ =0.01$. It has a single issue of zero-coupon debt that promises to pay $ \$8 $ $million$ in $T = 4$ years. If the continuously-compounded riskfree rate equals $r = 0.03$, what is this debt's promised yield to maturity, $y(0,T)$?

State your answer as a continuously-compounded percentage rate to 2 decimal places, e.g., $6.43$ .

```python
A0 = 10 * 10**6
σ = .4
δ = .01
r = .03
F = 8 * 10**6
T = 4 #years

d1 = (log(A0/F) + (r-δ+0.5*σ**2)*T)/(σ*sqrt(T))
d2 = d1 - σ*sqrt(T)

D0 = (exp(-r*T) * F * norm.cdf(d2) + exp(-δ*T)*A0*norm.cdf(-d1))/F
y0 = (-1/T) * log(D0)
print(y0*100, "%")
print(d1)
print(D0)
print(norm.cdf(d2))
print(norm.cdf(-d1))
```
